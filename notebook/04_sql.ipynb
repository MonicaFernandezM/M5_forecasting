# Conexión con SQL

```python
import pandas as pd 
import numpy as np
import pymysql
from sqlalchemy import create_engine
import getpass 
import mysql.connector
from sqlalchemy import text
```

```python
df = pd.read_csv("../processed/df_model_CA1.csv")
volatility = pd.read_csv("../processed/item_volatility.csv")
critical = pd.read_csv("../processed/top_critical_products.csv")
```

```python
#pip install mysql-connector-python
```

```python
bd = "retail_forecasting"
password = getpass.getpass("Introduce tu contraseña MySQL: ")
connection_string = f"mysql+pymysql://root:{password}@localhost/{bd}"
engine = create_engine(connection_string)
engine
```

```python
conn = mysql.connector.connect(host='localhost',
                        user='root',
                        passwd=password)
```

```python
cursor = conn.cursor()
```

```python
with engine.connect() as conn:
    result = conn.execute(text("SHOW TABLES;"))
    for row in result:
        print(row)
```

```python
df.to_sql(
    "df_model_CA1",
    engine,
    if_exists="replace",
    index=False
)
```

```python
volatility.to_sql(
    "item_volatility",
    engine,
    if_exists="replace",
    index=False
)
```

```python
critical.to_sql(
    'top_critical_products', 
    engine, 
    if_exists="replace",
    index=False
)
```
